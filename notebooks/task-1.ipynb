```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("../data/raw/bank_reviews.csv")
df.tail()
```

```python
# Rename and filter columns
df_cleaned = df[['content', 'score', 'at', 'bank', 'source']].copy()
df_cleaned.columns = ['review', 'rating', 'date', 'bank', 'source']

# Format date to YYYY-MM-DD
df_cleaned['date'] = pd.to_datetime(df_cleaned['date']).dt.date

# Drop duplicates and missing review text
df_cleaned = df_cleaned.drop_duplicates(subset=['review', 'bank'])
df_cleaned = df_cleaned.dropna(subset=['review'])

# Save to CSV (Remember to add data/ to .gitignore!)
df_cleaned.to_csv('../data/raw/cleaned_bank_reviews.csv', index=False)
print("Data saved successfully!")
```
